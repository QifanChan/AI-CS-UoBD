1. Given two arrays `a = np.array([1, 2, 3, 4])` and `b = np.array([5, 6, 7, 8])`, use a Numpy ufunc to compute the element-wise addition of `a` and `b`.

```python
import numpy as np

a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

print(a + b)
np.add(a, b)
```

17. Compute the ceiling value of each element in the array `u = np.array([-1.1, -0.9, 0, 0.9, 1.1])` using a Numpy ufunc.

```python
'''
The ceiling value, also known as the "ceiling function," rounds a number up to
the nearest integer that is greater than or equal to the original number.
In other words, it finds the smallest integer greater than or equal to the given number.
It's often denoted as ⌈x⌉.
'''
u = np.array([-1.1, -0.9, 0, 0.9, 1.1])
np.ceil(u)
```
