```python
# 9. What is the difference between `numpy.array()` and `numpy.asarray()`?
"""
    They both can convert input list into NumPy arrays,
    but
      1. np.array() creates a copy of input list,
      while np.asarray() creates a view of input list.
      if there the input is already a ndarray, np.asarray won't create a copy of it
      while np.array() will always create a copy of it
      2. np.array() can be used to cast the input data into a specific data type,
      while np.asarray() will infer data type from the input data.
"""
li = [1,2,3,4]
x = np.array(li)
print(x)
print(x.dtype)
print(type(x))

asarray = np.array(li)
x = np.asarray(asarray)
print(x)
print(x.dtype)
print(type(x))
```

```python
# 14. How do you create a diagonal matrix with specific diagonal values using numpy?
"""np.diag([])"""
# Create a list of diagonal values
diagonal_values = [1, 2, 3]

# Create a diagonal matrix from the list of diagonal values
diagonal_matrix = np.diag(diagonal_values)

# Print the diagonal matrix
print(diagonal_matrix)
```
