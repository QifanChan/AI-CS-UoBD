#### Question 30
Consider the following NumPy array, representing the annual average temperatures (in degrees Celsius) for 12 cities from January to December. Write the NumPy code to extract the temperatures for the first quarter (January to March) for City 5. Explain your indexing approach.

```python
temperatures = np.array([
    [5, 7, 10, 15, 20, 25, 30, 29, 24, 18, 12, 6],   # City 1
    [6, 9, 11, 16, 21, 26, 31, 30, 25, 19, 13, 7],   # City 2
    [8, 10, 13, 18, 23, 28, 33, 32, 27, 21, 15, 9],  # City 3
    [7, 9, 12, 17, 22, 27, 32, 31, 26, 20, 14, 8],   # City 4
    [5, 8, 11, 16, 21, 26, 31, 30, 25, 19, 13, 7],   # City 5
    [10, 12, 15, 20, 25, 30, 35, 34, 29, 23, 17, 11],# City 6
    [11, 13, 16, 21, 26, 31, 36, 35, 30, 24, 18, 12],# City 7
    [4, 7, 9, 14, 19, 24, 29, 28, 23, 17, 11, 5],    # City 8
    [6, 8, 11, 15, 20, 25, 30, 29, 24, 18, 12, 6],   # City 9
    [9, 11, 14, 19, 24, 29, 34, 33, 28, 22, 16, 10], # City 10
    [3, 5, 8, 13, 18, 23, 28, 27, 22, 16, 10, 4],    # City 11
    [4, 6, 8, 13, 18, 23, 28, 27, 22, 16, 10, 5]     # City 12
])
```

```python
# In Numpy, array index starts at 0, therefore, index 4 is City 5
# And we only need to extract the temeratures for 3 months (1 - 3)
# Use slice "up to but not include" [:3] to get the months at
# the index of 0, 1, 2 in the 2nd dimension
temperatures[4][:3] # Use the index of a 2D array
```

```python
# easier solution
temperatures[4, :3] # fancy index of numpy
```

#### Question 31
Using boolean indexing, select all the cities that have an average temperature above 20 degrees Celsius in July. Describe the steps you take to achieve this.

```python
"""
select all the cities that have an average temperature above 20 degrees Celsius in July,
it means we need to go through all rows, therefore, we use [:,] for rows.
And we only need to the July temperature is in the 7th column of each row, whose index is 6
So the final slice we use is [:, 6]
To create a boolean value we use comparsion "temperatures[:, 6] > 20" as a mask
Now we can put the mask in the index area temperatures[temperatures[:, 6] > 20]
"""
temperatures[temperatures[:, 6] > 20] # use the mask of numpy
```

#### Question 32
City 8 is planning an event and needs to know their average temperature for the months where City 3's temperature is below 15 degrees Celsius. Write the NumPy code to find this information and explain the logic behind your solution.

```python
# the months where City 3's temperature is below 15 degree
# temperatures[2, :] < 15

# For city 8, the outer index is temperatures[7, columns]
# columns = temperatures[2, :] < 15
# For the total average temperature, we can use np.mean()
# or ndarray.mean()
temperatures[7, temperatures[2, :] < 15].mean()
```

```python
city_3_cold_months = temperatures[2] < 15
city_8_temps_during_city_3_cold = temperatures[7, city_3_cold_months]
average_temp_city_8 = np.mean(city_8_temps_during_city_3_cold)
print("Average temperature for City 8 during City 3's cold months:", average_temp_city_8)
```
