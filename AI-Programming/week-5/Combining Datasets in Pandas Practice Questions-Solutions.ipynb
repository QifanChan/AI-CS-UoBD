# Combining Datasets in Pandas Coding Practice Questions

1. Create two DataFrames `df1` and `df2` with the following data:
   ```
   df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                      'B': ['B0', 'B1', 'B2', 'B3'],
                      'key': ['K0', 'K1', 'K2', 'K3']})
   
   df2 = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3'],
                      'key': ['K0', 'K1', 'K2', 'K3']})
   ```
   Display the DataFrames.

2. Merge `df1` and `df2` on the 'key' column using an inner join. Display the result.

3. Merge `df1` and `df2` on the 'key' column using an outer join. Display the result.

4. Merge `df1` and `df2` on the 'key' column using a left join with `df1` as the left DataFrame. Display the result.

5. Merge `df1` and `df2` on the 'key' column using a right join with `df2` as the right DataFrame. Display the result.

6. Concatenate `df1` and `df2` along the rows. Display the result.

7. Concatenate `df1` and `df2` along the columns. Display the result.

8. Use the `append` method to add `df2` to `df1`. Display the result.

9. Create a MultiIndex DataFrame `df3` using the following data:
   ```
   df3 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                      'B': ['B4', 'B5', 'B6', 'B7'],
                      'key': ['K0', 'K1', 'K2', 'K3'],
                      'key2': ['K4', 'K5', 'K6', 'K7']})
   ```
   Display the DataFrame.

10. Merge `df1` and `df3` using multiple keys: 'key' and 'key2'. Display the result.

11. Use the `join` method to combine `df1` and `df2` using the 'key' column. Display the result.

12. Use the `pivot` method to reshape `df1` such that 'A' and 'B' columns become columns and the 'key' column becomes the index. Display the result.

13. Create a pivot table from `df1` using the 'key' column as the index and 'A' as values. Display the result.

14. Use the `melt` method to reshape `df1` to long format. Display the result.

15. Use the `stack` method to reshape `df1`. Display the result.

16. Use the `unstack` method to reshape the stacked `df1`. Display the result.

17. Use the `crosstab` method to compute a cross-tabulation of the 'A' and 'B' columns from `df1`. Display the result.

18. Use the `pivot_table` method to create a pivot table from `df1` using 'A' as values and 'key' as the index. Display the result.

19. Use the `groupby` method to group `df1` by the 'key' column and compute the mean of the 'A' column. Display the result.

20. Use the `agg` method to compute multiple aggregations on `df1`: mean of 'A' and sum of 'B'. Display the result.

# Solutions to Combining Datasets in Pandas Coding Practice Questions

````python
import pandas as pd
import numpy as np
````

````python
# Solution to Question 1
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                   'B': ['B0', 'B1', 'B2', 'B3'],
                   'key': ['K0', 'K1', 'K2', 'K3']})

df2 = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                   'D': ['D0', 'D1', 'D2', 'D3'],
                   'key': ['K0', 'K1', 'K2', 'K3']})
df1, df2
````

````python
# Solution to Question 2
merged_inner = pd.merge(df1, df2, on='key', how='inner')
merged_inner
````

````python
# Solution to Question 3
merged_outer = pd.merge(df1, df2, on='key', how='outer')
merged_outer
````

````python
# Solution to Question 4
merged_left = pd.merge(df1, df2, on='key', how='left')
merged_left
````

````python
# Solution to Question 5
merged_right = pd.merge(df1, df2, on='key', how='right')
merged_right
````

````python
# Solution to Question 6
concatenated_rows = pd.concat([df1, df2], axis=0)
concatenated_rows
````

````python
# Solution to Question 7
concatenated_columns = pd.concat([df1, df2], axis=1)
concatenated_columns
````

````python
# Solution to Question 8
appended_df = df1.append(df2)
appended_df
````

````python
# Solution to Question 9
df3 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                   'B': ['B4', 'B5', 'B6', 'B7'],
                   'key': ['K0', 'K1', 'K2', 'K3'],
                   'key2': ['K4', 'K5', 'K6', 'K7']})
df3
````

````python
# Solution to Question 10
merged_multi_keys = pd.merge(df1, df3, on=['key', 'key2'], how='inner')
merged_multi_keys
````

````python
# Solution to Question 11
joined_df = df1.join(df2.set_index('key'), on='key')
joined_df
````

````python
# Solution to Question 12
pivoted_df = df1.pivot(index='key', columns='A', values='B')
pivoted_df
````

````python
# Solution to Question 13
pivot_table_df = df1.pivot_table(index='key', values='A', aggfunc='first')
pivot_table_df
````

````python
# Solution to Question 14
melted_df = df1.melt(id_vars='key', value_vars=['A', 'B'])
melted_df
````

````python
# Solution to Question 15
stacked_df = df1.stack()
stacked_df
````

````python
# Solution to Question 16
unstacked_df = stacked_df.unstack()
unstacked_df
````

````python
# Solution to Question 17
cross_tab_df = pd.crosstab(df1['A'], df1['B'])
cross_tab_df
````

````python
# Solution to Question 18
pivot_table_agg_df = df1.pivot_table(index='key', values='A', aggfunc=[np.mean, sum])
pivot_table_agg_df
````

````python
# Solution to Question 19
grouped_df = df1.groupby('key')['A'].mean()
grouped_df
````

````python
# Solution to Question 20
aggregated_df = df1.agg({'A': 'mean', 'B': 'sum'})
aggregated_df
````
