# Numpy Aggregate Functions Coding Lab Questions

1. Create a 1-D numpy array of 10 random integers and compute the sum of all its elements.

2. Given a 2-D numpy array of shape (5, 5), compute the minimum value of each row.

3. Compute the cumulative sum of a 1-D numpy array of 10 elements.

4. Find the mean value of a 2-D numpy array of shape (4, 4).

5. Compute the standard deviation of a 1-D numpy array of 20 elements.

6. Given a 2-D numpy array of shape (3, 3), compute the maximum value of each column.

7. Compute the median of a 1-D numpy array of 15 elements.

8. Find the variance of a 2-D numpy array of shape (5, 3).

9. Compute the product of all elements in a 1-D numpy array of 10 elements.

10. Given a 2-D numpy array of shape (6, 4), compute the cumulative product along each row.

11. Compute the 25th percentile of a 1-D numpy array of 50 elements.

12. Find the 75th percentile of a 2-D numpy array of shape (7, 7).

13. Compute the sum of all elements in a 2-D numpy array of shape (3, 3) along the diagonal.

14. Given a 1-D numpy array of 20 elements, compute the sum of its positive elements.

15. Compute the mean value of each column in a 2-D numpy array of shape (4, 5).

16. Find the minimum value of a 2-D numpy array of shape (5, 6) along the diagonal.

17. Compute the standard deviation of each row in a 2-D numpy array of shape (3, 7).

18. Given a 1-D numpy array of 30 elements, compute the difference between its maximum and minimum values.

19. Compute the product of all non-zero elements in a 1-D numpy array of 25 elements.

20. Find the variance of each column in a 2-D numpy array of shape (8, 4).

# Solutions

```python
import numpy as np
```

```python
# Solution to Question 1
array_1d = np.random.randint(0, 100, 10)
sum_1 = np.sum(array_1d)
sum_1
```

```python
# Solution to Question 2
array_2d = np.random.randint(0, 100, (5, 5))
min_values = np.min(array_2d, axis=1)
min_values
```

```python
# Solution to Question 3
array_1d = np.random.randint(0, 100, 10)
cumsum_3 = np.cumsum(array_1d)
cumsum_3
```

```python
# Solution to Question 4
array_2d = np.random.randint(0, 100, (4, 4))
mean_4 = np.mean(array_2d)
mean_4
```

```python
# Solution to Question 5
array_1d = np.random.randint(0, 100, 20)
std_5 = np.std(array_1d)
std_5
```

```python
# Solution to Question 6
array_2d = np.random.randint(0, 100, (3, 3))
max_values = np.max(array_2d, axis=0)
max_values
```

```python
# Solution to Question 7
array_1d = np.random.randint(0, 100, 15)
median_7 = np.median(array_1d)
median_7
```

```python
# Solution to Question 8
array_2d = np.random.randint(0, 100, (5, 3))
variance_8 = np.var(array_2d)
variance_8
```

```python
# Solution to Question 9
array_1d = np.random.randint(1, 10, 10)
product_9 = np.prod(array_1d)
product_9
```

```python
# Solution to Question 10
array_2d = np.random.randint(1, 10, (6, 4))
cumprod_10 = np.cumprod(array_2d, axis=1)
cumprod_10
```

```python
# Solution to Question 11
array_1d = np.random.randint(0, 100, 50)
percentile_25 = np.percentile(array_1d, 25)
percentile_25
```

```python
# Solution to Question 12
array_2d = np.random.randint(0, 100, (7, 7))
percentile_75 = np.percentile(array_2d, 75)
percentile_75
```

```python
# Solution to Question 13
array_2d = np.random.randint(0, 100, (3, 3))
diagonal_sum = np.trace(array_2d)
diagonal_sum
```

```python
# Solution to Question 14
array_1d = np.random.randint(-50, 50, 20)
positive_sum = np.sum(array_1d[array_1d > 0])
positive_sum
```

```python
# Solution to Question 15
array_2d = np.random.randint(0, 100, (4, 5))
mean_values = np.mean(array_2d, axis=0)
mean_values
```

```python
# Solution to Question 16
array_2d = np.random.randint(0, 100, (5, 6))
diagonal_min = np.min(np.diagonal(array_2d))
diagonal_min
```

```python
# Solution to Question 17
array_2d = np.random.randint(0, 100, (3, 7))
std_values = np.std(array_2d, axis=1)
std_values
```

```python
# Solution to Question 18
array_1d = np.random.randint(0, 100, 30)
diff_18 = np.max(array_1d) - np.min(array_1d)
diff_18
```

```python
# Solution to Question 19
array_1d = np.random.randint(0, 10, 25)
product_19 = np.prod(array_1d[array_1d != 0])
product_19
```

```python
# Solution to Question 20
# Find the variance of each column in a 2-D numpy array of shape (8, 4).
array_2d = np.random.randint(0, 100, (8, 4))
variance_values = np.var(array_2d, axis=0)
variance_values
```
