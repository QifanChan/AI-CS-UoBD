1. Given two arrays `a = np.array([1, 2, 3])` and `b = np.array([1, 2])`, demonstrate how broadcasting would work between these two arrays if you try to add them.

```python
import numpy as np
```

```python
# my answer - wrong
a = np.array([1, 2, 3])
b = np.array([1, 2])
b = b[:, np.newaxis]
a + b
```

```python
# right answer
a = np.array([1, 2, 3])
b = np.array([1, 2])
try:
    result = a + b
    print(result)
except ValueError as e:
    print("operands could not be broadcast together with shapes (3,) (2,)")
```
