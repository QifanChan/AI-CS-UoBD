#### Logical Operators

```python
# and - True only if both operands are True, otherwise false
print((5 == 5) and (1 == 2))
print((5 == 5) and (1 != 1))

# or - True if anyone of its operands is True. False if both operands are False.
print((5 == 5) or (1 == 2))
print((5 > 5) or (1 != 1))
```

#### Math functions
Python's ecosystem is built around packages and modules. A package/module is usually a set of specialised functions in a particular domain of interest.

We are going to _use_ many packages throughout this module.

> In Python, both modules and packages organize and structure the code but serve different purposes. A module is a single file containing python code, whereas a package is a collection of modules that are organized in a directory hierarchy.

To use a module, we need to `import` it. It informs Python about all functions available in a module.

```python
import math     # math is one of the thousands of packages available in Python
dir(math)       # dir is used to see what are available functions and variables in a package
```

#### Strings
Represented by alphanumeric sequence within $["~"]$ or $['~~']$
Strings have different properties or functions, e.g., `len` is property; `tolower()` is a function.

```python
hw = "Hello World"
len_hw = len(hw)          # len is a function to find the length of an object in Python
print("length of the string is:", len_hw)   # uses the variable
print("length of the string is:", len(hw))  # uses the function directly. Does not involve any variable
```

They are like lists but:
1. they are not mutable: no insertion/deletion
2. they use ( ) instead of [ ]
3. element access in tuples is faster

```python
tuplea = (1, 2, 3)
tuplea
```

### Sets
 An unordered collection with no duplicate elements. It is also support the mathematical operations like union, intersection, difference, and symmetric difference.

 Can be used to find unique elements in a string, list, etc.

```python
seta = {"h", "e", "l", "l", "o"}
seta
```

```python
setb = set(list("hollow"))
setb
```

```python
print(seta.union(setb))
print(seta.intersection(setb))
print(seta.difference(setb))
print(setb.difference(seta))
# a number of other methods
#
```

```python
print(len(tuplea), len(lista), len(seta))
print(type(tuplea), type(lista), type(seta))
```
